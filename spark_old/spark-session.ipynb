```python
import pandas as pd
```

```python
df = pd.read_csv("../stock-market/ADANIPORTS.csv")
df.head()
```

```python
df["Date"] = pd.to_datetime(df['Date'])
df.info()
```

```python
df["Year"] = df["Date"].apply(lambda x: x.year)
df["Month"] = df["Date"].apply(lambda x: x.month)
df["Day"] = df["Date"].apply(lambda x: x.day)

df.head()
```

```python
a = df.groupby(["Year", "Month"]).agg(["min", "max"])
a["High"]["min"]
```
